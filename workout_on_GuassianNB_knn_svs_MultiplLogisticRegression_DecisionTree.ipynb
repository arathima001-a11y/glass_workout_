```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, ConfusionMatrixDisplay
```

```python
df1 = pd.read_csv("glass.csv")
```

```python
X = df1.drop('Type', axis=1)
y = df1['Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
gb=GaussianNB()
knn=KNeighborsClassifier(n_neighbors=3)
svc=SVC(kernel='linear')
```

```python
lst=[gb,knn,svc]
for i in lst:
    print(i,"====>")
    i.fit(X_train,y_train)
    
    y_train_pred=i.predict(X_train)
    y_test_pred=i.predict(X_test)

    print("\nTrain Accuracy:",(accuracy_score(y_train, y_train_pred) * 100))
    print("\nTest Accuracy:",(accuracy_score(y_test, y_test_pred) * 100),"\n")
    print("-------------------------------------------------------------------\n\n")
```
